```python
#Visualizations 
#check balanced or imbalanced the  data set
df.quality.value_counts().plot(kind="bar")#shows the count of each unique value in the quality column as a bar chart
plt.xlabel("Wine Quality")#sets the label for the x-axis of the plot to "Wine Quality"
plt.ylabel("Count")#sets the label for the y-axis of the plot to "Count"
plt.show()#displays the plot on the screen
```

```python
for column in df.columns:
   sns.histplot(df[column], kde=True)#creates a histogram with a kernel density estimate (KDE) for each column in the dataset, which helps to visualize the distribution of the data in each column. The KDE adds a smooth curve to the histogram, making it easier to see the overall shape of the distribution.
```

```python
#it has to be curved distribution ,above have no curve distribution,curve has broken

sns.histplot(df["alcohol"], kde=True)#creates a histogram with a kernel density estimate (KDE) for the "alcohol" column in the dataset, which helps to visualize the distribution of the data in that column. The KDE adds a smooth curve to the histogram, making it easier to see the overall shape of the distribution.
```

```python
sns.pairplot(df, hue="quality")#creates a pair plot of the dataset, which shows the relationships between all pairs of numerical columns in the dataset, with different colors representing different values of the "quality" column. This helps to visualize how the different features in the dataset are related to each other and to the target variable (quality).
#this use to check the relationship between the features and the target variable (quality) and to identify any potential patterns or trends in the data. 
# univariate,bivariate, multivariate analysis .
# univariate analysis : it is the analysis of a single variable, it is used to understand the distribution of the data and to identify any potential outliers in the data.
# bivariate analysis : it is the analysis of two variables, it is used to understand the
# multivariate analysis : it is the analysis of more than two variables, it is used to understand the relationship between multiple variables and to identify any potential patterns or trends in the data. 
```

```python
sns.catplot(x="quality", y="alcohol", data=df, kind="box")#creates a box plot of the "alcohol" column grouped by the "quality" column, which helps to visualize the distribution of alcohol content for each quality level and identify any potential outliers in the data.
```
