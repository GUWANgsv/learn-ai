### EDA & feature engineering

#### feature is an individual measurable property or characteristic of a phenomenon being observed.  (hoya gana data set ekaka property eg ex: age, height, weight, etc column=> feature,row=> value)

#### feature engineering the process of using domain knowledge to create new features from raw data.


#### EDA (Exploratory Data Analysis) is the process of analyzing and visualizing data to understand its main characteristics, often using statistical graphics and other data visualization methods.
main characteristics of data include:
1. Distribution: Understanding how the data is distributed (eg: normal distribution, skewed distribution, etc)
2. Central Tendency: Measures of central tendency (eg: mean, median, mode) to understand the typical value in the data
3. Variability: Measures of variability (eg: range, variance, standard deviation) to understand how spread out the data is
4. Outliers: Identifying any outliers in the data that may affect the analysis
5. Relationships: Exploring relationships between different features (eg: correlation, scatter plots, etc)

#### Handling Numberical data
1. Scaling/Normalization: ensuring all numbers are on a similar scale (eg: bringing both "age" and "income" to the same scale like range 0-1)
2. Binning : Converting continuous numbers into groups  (eg: age groups-> 0-18='Minor', 19-60='Adult')

#### Dealing with missing data
1. Imputation: filling in missing values with mean, median, or mode depending on the type of data
mean/median--> numerical data
mode--> categorical data
2. Dropping:Some times if there are too many missing values in a feature, it might be best to drop that feature entirely (eg: if 80% of the "income" data is missing, we might choose to drop the "income" column)
 
#### feature creation & Extraction
1. Polynomial features: Combining existing features to create new ones (eg: length and width of a rectangle can be combined to create a new feature called "area")
2. Date/Time features: Extracting useful information from date and time data (eg: from a "date of birth" feature, we can extract "age" or "day of the week")
3. Log transformation: Applying a logarithmic transformation to skewed data to make it more normally distributed (eg: if "income" is heavily skewed, we can apply a log transformation to reduce the skewness)
#### feature selection
Too many features can lead to overfitting ( occurs when a machine learning model learns training data noise and details too closely, resulting in high accuracy on training data but poor performance on unseen data)
Techniques for feature selection:
* Correlation Matrix:Removing features that are too similar to each other 
* Feature Importance: Using models to tell us which features they actually care about

\[ note : Underfitting occurs when a machine learning model is too simple to capture the underlying patterns in data, resulting in poor performance on both training and test datasets ]
