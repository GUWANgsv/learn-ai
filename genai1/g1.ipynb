## NLP Basic
- Tokenization:  
 Breaking text into smaller units (tokens) such as words or sentences.
- Stemming and Lemmatization:  
 Reducing words to their base or root form.
- Stop Words Removal:  
 Removing common words that do not carry significant meaning (e.g., "the," "is," "and").
(Tokenizer, Stemmer, Lemmatizer, StopWordsRemover  this 4 is pre processing step in NLP)
- vectorization:   
Converting text into numerical representations (vectors) that can be processed by machine learning algorithms.
- Word Embeddings:   
Representing words in a continuous vector space where semantically similar words are closer together (e.g.  BoW, GloVe, TF IDE and Word2Vec).

### Stop Words Removal
- Removing common words that do not carry significant meaning (e.g., "the," "is," "and") to reduce noise in text data.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "Natural language processing (NLP) is a field of artificial intelligence that enables computers to understand and process human language. It combines computational linguistics with machine learning, deep learning, and statistical modeling to analyze and generate human language. NLP is used in various applications such as chatbots, sentiment analysis, language translation, and information retrieval."
print("Original text:", text)

# Tokenize the text into words
words = word_tokenize(text)
print("print toknize words:")
print(words)
print(" \n")
#Case folding - converting all characters to lowercase
text = text.lower()
print("Case folding:", text)

words=word_tokenize(text)
print("print toknize words:")
print(words)
print(" \n")
print(" \n")
# Remove stop words
filtered_words = [word for word in words if word not in stopwords.words('english')]
print("Filtered words:")
print(filtered_words)

removed_words=[word for word in words if word  in stopwords.words('english')]
print("Removed words:")
print(removed_words)
#for word in words → Loop through every word in the list words.
#if word in stopwords.words('english') → Check whether that word exists in the English stopword list
#[ ... ] → Create a new list from all words that pass the condition.
```

### Stemming and Lemmatization
Reducing words to their base form.
#### Stemming

Definition: Cuts off prefixes/suffixes to get a root form.

Method: Rule-based chopping, no understanding of grammar or meaning.

Output: Sometimes not a real word.

Examples

- playing → play
- studies → studi
- better → better (unchanged)

Stemming is fast but rough.
#### Lemmatization

Definition: Converts a word to its correct dictionary base form (lemma).

Method: Uses vocabulary + grammar rules.

Output: Always a real word.

Examples

- playing → play
- studies → study
- better → good

Lemmatization is slower but more accurate.


```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
#initialize the Porter Stemmer
stemmer=PorterStemmer()
words=["big","bigger","biggest","running","runner","ran","runs","easily","fairly","studies","studying","studied","studies","studious","studiously","studiousness","studiousnesses","studiousnesses","happily","happiness","happier","happiest"]
print("Original words:")
print(words)
print(" \n")
#Apply stemming technique
stemmed_words=[stemmer.stem(word) for word in words]
print("Stemmed words:")
print(stemmed_words)

#intialize the WordNet Lemmatizer
lemmatizer=WordNetLemmatizer()
#Apply lemmatization technique
lemmatized_words=[lemmatizer.lemmatize(word) for word in words]
print("Lemmatized words:")
print(lemmatized_words)
```

##### Term Frequency-Inverse Document Frequency (TF-IDF)

*  TF=(number of times the word appears in a document/total number of words in the document)
* IDF=(log(total number of documents/number of documents containing the word))
* TF-IDF=TF*IDF

[s1,s2,s3]=>3 documents \

- TF(boy)=1/2=0.5
- TF(girl)=0/2=0

- TF(good)=1/2=0.5
- TF(boy)=0/2=0
- TF(girl)=1/2=0.5

s3:
- TF(good)=1/3=0.33
- TF(boy)=1/3=0.33
- TF(girl)=1/3=0.33

IDF 
- IDF(good)=log(3/3)=0
- IDF(boy)=log(3/2)=0.18
- IDF(girl)=log(3/2)=0.18

Think of fit_transform as creating a dictionary and then translating a book using that dictionary.
The Simple Analogy:
Imagine you are teaching a computer to read Spanish.
1. 2Training Phase (fit_transform):
- You give it the Training Book.
- Fit: The computer reads the book and makes a dictionary of every Spanish word it finds (e.g., 5,000 words).
- Transform: It replaces those words with numbers so the model can understand them.
2. Testing Phase (Only transform):
- Now you give it a New Book (Test Set).
- If you use fit_transform again, the computer throws away the old dictionary and makes a new one. If this book only has 3,000 words, the "page numbers" (columns) won't match the first dictionary.
- The Result: The computer gets confused because "Word #10" in the first book was "Hola," but in the second book, "Word #10" is "Gato."
Why only one time?
Consistency: You need "Word #10" to mean the same thing in every dataset you ever show your model.
The Shape: Your model is built like a machine with a specific number of "slots" (e.g., 7,739). If you change the dictionary, you change the number of slots, and the data won't fit into the machine anymore.

* Example:

##### 6. K-Nearest Neighbors (KNN)

* Example:

##### TF-IDF practical

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
    "Apple is announcing new iphone tomorrow",
    "Tesla is announcing new model-3 tomorrow",
    "Google is announcing new pixel-6 tomorrow",
    "Microsoft is announcing new surface tomorrow",
    "Amazon is announcing new eco-dot tomorrow",
    "I am eating biryani and you are eating grapes"
]
```

```python
vectorizer1=TfidfVectorizer()
vectorizer1.fit(corpus)
transform_output=vectorizer1.transform(corpus)
```

```python
print(vectorizer1.vocabulary_)
```

```python
i = vectorizer1.vocabulary_.get('thor')
vectorizer1.idf_[i]
```

```python
# Print the idf of each word

all_feature_names = vectorizer1.get_feature_names_out()

for word in all_feature_names:

  indx = vectorizer1.vocabulary_.get(word)

  #get the score
  idf_score = vectorizer1.idf_[indx]

  print(f"{word}: {idf_score}")
```

##### TF-IDF use case

```python
df = pd.read_csv('Ecommerce_data.csv')
df.head(5)
```

```python
df.label.value_counts()
```

```python
df.shape
```

```python
df['label'] = df['label'].map({
    'Household': 0,
    'Electronics': 1,
    'Clothing & Accessories': 2,
    'Books': 3
})
df.head(5)
```

```python
#Train Test Split
X_train, X_test, y_train, y_test = train_test_split(df.Text, df.label, test_size=0.2)
```

```python

len(X_test)
```

```python
len(X_train)
```

```python
# Tfidf Vectorizer
tf = TfidfVectorizer()

X_train_tf = tf.fit_transform(X_train)
X_test_tf = tf.transform(X_test)
```

```python
# Classification Model
clf = DecisionTreeClassifier()
clf.fit(X_train_tf,y_train)

y_pred = clf.predict(X_test_tf)
```

```python
print(classification_report(y_test, y_pred))
```
