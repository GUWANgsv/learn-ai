## linear regression
Linear regression is a fundamental statistical method used to model the relationship between a dependent variable and one or more independent variables. The goal of linear regression is to find the best-fitting straight line (or hyperplane in higher dimensions) that describes how the dependent variable changes as the independent variables change.

types of linear regression:
1. Simple Linear Regression: Involves one independent variable and one dependent variable. 
2. Multiple Linear Regression: Involves two or more independent variables and one dependent variable.
 

y=w.x+b

y: dependent variable (target)  

x: independent variable (feature)

w: weight (coefficient)

b: bias (intercept)
in regression, the main goal is to find the best-fit line to the data.

A good model usually has:

✅ R²
High (closer to 1)

✅ Errors (MSE, RMSE, MAE)

* Low
* Consistent across train & test data

---

#### 🚨 VERY IMPORTANT (Most people miss this)

 1. Compare Train vs Test

If:

* Train R² = 0.95
* Test R² = 0.60

👉 Model is **overfitting → NOT reliable**



 2. Domain matters

* Finance → small errors required
* Weather → larger errors acceptable
