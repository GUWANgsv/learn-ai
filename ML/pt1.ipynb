# ML
ML TYPES
- Supervised Learning
- Unsupervised Learning
- Reinforcement Learning


## supervised learning

goal: predict the output of new data based on the training data or historical data

data:Includes (Independent features) inputs and (Dependent feature) output (labeled output).

types:
- Regression: output is a continuousvalue (e.g., house price, stock price)
- Classification: output is a discrete value (e.g., spam or not spam, cat or dog)

## unsupervised learning

goal: find hidden patterns or structures/groupings 

data:Includes onlyinputs, no labeled/target  output to predict.


## reinforcement learning
goal: learn to make a sequence of decisions by interacting with an environment to maximize cumulative reward.

data:Includes (State) the current situation of the environment, (Action) the decision made by the agent, and (Reward) feedback received from the environment after taking an action.






use case-->ML model-->instance based or model based


Instance-Based Learning: "The Memorizer"
This student doesn't try to understand the "why"—they just memorize every single practice question and answer.
How it works: When they see a new question on the exam, they look through their memory for the question that looks the most similar to it and give the same answer.



Model-Based Learning: "The Formula Finder"
This student looks at the practice questions to find a pattern or a general rule
How it works: Once they have the formula, they throw away the practice books. On the exam, they just plug the new numbers into their formula to get the answer.
